```python
#Read in the CSV as a DataFrame
netflix_df = pd.read_csv("C:/Users/user/Documents/Data Analysis Course/CWW AFRICA\Individual project/CWW Tech 4.0 Individual Project/datasets/netflix_data.csv")

#Print the first five rows of the DataFrame
print(netflix_df.head(5))
```

```python
#Subset the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

#Select only the columns of interest
netflix_movies_col_subset = netflix_df_movies_only.loc[:, ['title', 'country', 'genre', 'release_year', 'duration']]

#Print the first five rows of the new DataFrame
print(netflix_movies_col_subset.head(5))

```

```python
#Create a figure and increase the figure size
fig = plt.figure(figsize = (12,8))

#Create a scatter plot of duration versus year
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'])

#Create a title
plt.title('SCATTER PLOT FOR DURATION VERSUS YEAR')

#Show the plot
plt.show()
```

```python
#Filter for durations shorter than 60 minutes
short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration'] < 60]

#Print the first 20 rows of short_movies
print(short_movies.head(20))

```

```python
# Define an empty list
colors = []

# Iterate over rows of netflix_movies_col_subset
for index, row in netflix_movies_col_subset.iterrows() :
    genre = row['genre']
    if genre == 'Children':
        colors.append ('red')
    elif genre == 'Stand-Up' :
        colors.append ('blue')
    elif genre == 'Documentaries':
        colors.append('Green')
    else:
        colors.append('Black')
        
# Inspect the first 10 values in your list 
print(colors[:10])
```

```python
# Set the figure style and initalize a new figure
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], color=colors, marker='o')

# Create a title and axis labels
plt.xlabel('Release_year')
plt.ylabel('Duration (minutes)')
plt.title('SCATTER PLOT OF DURATION VERSUS RELEASE_YEAR')

# Show the plot
plt.show()
```
